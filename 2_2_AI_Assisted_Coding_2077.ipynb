```python
def extract_negative_numbers(numbers):
    negative_numbers = []
    for num in numbers:
        if num < 0:
            negative_numbers.append(num)
    return negative_numbers

# Example usage:
my_list = [1, -2, 3, -4, 0, -5, 6]
negatives = extract_negative_numbers(my_list)

print(f"Original list: {my_list}")
print(f"Negative numbers: {negatives}")
```

```python
def separate_numbers(numbers):
    if not numbers:
        print("The given list is empty")
        return [], []

    negative_numbers = [num for num in numbers if num < 0]
    non_negative_numbers = [num for num in numbers if num >= 0]

    return negative_numbers, non_negative_numbers

# Get user input for a list of numbers
user_input = input("Enter numbers separated by spaces (e.g., 1 -2 3 -4 0 -5 6): ")

# Convert the input string to a list of integers
try:
    original_numbers = [int(num) for num in user_input.split()]
except ValueError:
    print("Invalid input. Please enter numbers separated by spaces.")
    original_numbers = [] # Set to empty list if input is invalid

# Call the function to separate numbers
negatives, non_negatives = separate_numbers(original_numbers)

# Print results if the list was not empty
if original_numbers:
    print(f"Original list from user input: {original_numbers}")
    print(f"Separated negative numbers: {negatives}")
    print(f"New list with no negative numbers: {non_negatives}")
```

```python
def count_chars(input_string):
    vowels = "aeiou"
    vowel_count = 0
    consonant_count = 0
    digit_count = 0

    if not input_string:
        print("Given String is empty")
        return

    for char in input_string:
        char_lower = char.lower()
        if 'a' <= char_lower <= 'z': # Check if it's an alphabet
            if char_lower in vowels:
                vowel_count += 1
            else:
                consonant_count += 1
        elif '0' <= char <= '9': # Check if it's a digit
            digit_count += 1

    print(f"Original String: {input_string}")
    if vowel_count == 0:
        print("No vowels are found.")
    else:
        print(f"Number of Vowels: {vowel_count}")

    if consonant_count == 0:
        print("No consonants are found.")
    else:
        print(f"Number of Consonants: {consonant_count}")

    if digit_count == 0:
        print("No digits are found.")
    else:
        print(f"Number of Digits: {digit_count}")

# Take input from the user
user_string = input("Enter a string: ")

# Call the function with the user's string
count_chars(user_string)
```

```python
import re

def check_palindrome(input_string):
    # Remove non-alphanumeric characters and convert to lowercase
    processed_string = re.sub(r'[^a-zA-Z0-9]', '', input_string).lower()

    # Check if the processed string is equal to its reverse
    if processed_string == processed_string[::-1]:
        return "yes! given string is a palindrome"
    else:
        return "no! the given string is not a palindrome"

# Take input from the user
user_input_string = input("Enter a string to check if it's a palindrome: ")

# Call the function and print the result
result = check_palindrome(user_input_string)
print(result)
```
